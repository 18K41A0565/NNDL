```
import numpy as np
import sympy as sp
a=np.random.randint(1,5)
print("\n The value of x generated randomly is ",a)
lr=0.01
max=50
count=0
x=sp.Symbol('x')
equation=x**4+3*x**2+10
expression=sp.diff(equation,x)
print("expression is ",expression)
equation1=lambda x:4*x**3+6*x
while  count < max :
  previous=a
  delta=-(lr)*equation1(previous)
  a=a+delta
  count=count+1
  print("Iteration",count)
  print("X value is{:.2f} ".format(a))
print("The local minimum occurs at {:.2f} ".format(a))
expression=expression.subs(x,a)
print(expression)
```
